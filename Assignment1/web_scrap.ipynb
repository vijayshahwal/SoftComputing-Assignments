```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
source = requests.get('https://www.goibibo.com/destinations/intl/all-countries/').text
source
```

```python
soup = BeautifulSoup(source,"html.parser")
cards = soup.find_all('div',{'class':'col-md-4 col-sm-4 col-xs-12 filtr-item posrel'})
country_urls = []
country_names = []
#print(cards[0])
#print(cards.find('a'))
for card in cards:
	for a in card.find_all('a', href=True):
		if a['href']:
			country_urls.append(a['href'])
			country_names.append(a.text)
print(country_urls)
print(country_names)
```

```python
length = len(country_urls)
city_urls = {}
for i in range(length):
	url = country_urls[i]
	country_name = country_names[i]
	data = requests.get(url)
	html = data.text
	soup = BeautifulSoup(html, 'html.parser')
	places_to_visit = soup.find('div', {'class' : 'place-to-visit-container'})
	if(places_to_visit):
		card = places_to_visit.find('div', {'class' : 'col-md-12'})
		for a in card.find_all('a', href=True):
			if a['href']:
				list = a['href'].split('/')
				city_urls[list[3]] = 'https://www.goibibo.com/hotels/intl-hotels-in-' + list[3] + '-ct/'
city_list =[]
for c in city_urls:
  city_list.append(c)
city_list
```

```python
hotel_price = {}
hotel_url = {}
cards_price_data=""
lis1=[]
lis2=[]
for city in city_urls:  
    #print(city)
    city_name = city
    #getting all hotels in a city along with their price
    url = city_urls[city]
    response = requests.get(url)
    data = BeautifulSoup(response.text, 'html.parser')
    cards_price_data = data.find_all('div',attrs={'class','HotelCardstyles__CurrentPriceTextWrapper-sc-1s80tyk-30 daRNvo'})
    cards_url_data = data.find_all('div', attrs={'class','HotelCardstyles__HotelNameWrapperDiv-sc-1s80tyk-15 htvOTn'})
    lis1.append(cards_price_data)
    lis2.append(cards_url_data)
    #print(city)
    #print(cards_price_data)
    for i in range(0,len(cards_price_data)):
      hotel_price[cards_url_data[i].text.rstrip()] = cards_price_data[i].text.rstrip()
      hotel_url[cards_url_data[i].text.rstrip()] = 'https://www.goibibo.com' + cards_url_data[i].find('a', href = True)['href']
print(hotel_price)
print(hotel_url)
print(len(lis1))
print(len(hotel_url))
```

```python
rating_list =[]
prince_list =[]
Amenities_list=[]
HotelDescription_list=[]
Review1_list=[]
Review2_list=[]
hotel_name_list=[]
for p in lis2:
  for i in range(0, len(p)):
      url = hotel_url[p[i].text.rstrip()]
      data = requests.get(url)
      html = data.text
      hotel_name = p[i].text.rstrip()
      pricel = hotel_price[p[i].text.rstrip()]

      soup = BeautifulSoup(html, 'html.parser')
      div = soup.find('div', { 'id': 'root' })
      description = div.find('section', {'class' : "HotelDetailsMain__HotelDetailsContainer-sc-2p7gdu-0 hcyUbD"})
      address = description.find('span', {'itemprop' : 'streetAddress'}).text.rstrip().replace(' View on Map', '')

      descriptionl = address

      rating = 'Rating not found'
      ratingdata = description.find('span', {'itemprop' : 'ratingValue'}) #contains rating
      if ratingdata:
        rating = ratingdata.text.rstrip()
      ratingl = rating

      review1 = 'Review not found'
      review2 = 'Review not found'
      reviews = div.find_all('span', {'class' : 'UserReviewstyles__UserReviewTextStyle-sc-1y05l7z-4 cHVFbF'})
      if(len(reviews) > 1):
        review1 = reviews[0].text.rstrip()
      if(len(reviews) > 3):
        review2 = reviews[3].text.rstrip()
      review1l = review1
      review2l = review2

      amenities_list = []  #contains all the amenities.
      amenitiesdiv = div.find('div', {'class' : 'Amenitiesstyles__AmenitiesListBlock-sc-10opy4a-5 gyoAXQ'})
      if(amenitiesdiv==None):
        continue
      for amenitiy in amenitiesdiv.find_all('span', {'class':'Amenitiesstyles__AmenityItemText-sc-10opy4a-9 eLihWE'}) :
        amenities_list.append(amenitiy.text.rstrip())
      # print('Amenities : ', end = '')
      amenities = amenities_list
      # print(amenities_list)
      hotel_name_list.append(hotel_name)
      rating_list.append(ratingl)
      prince_list.append(pricel)
      Amenities_list.append(amenities)
      HotelDescription_list.append(descriptionl)
      Review1_list.append(review1l)
      Review2_list.append(review2l)
print(len(hotel_name_list))
print(rating_list)
print(prince_list)
print(HotelDescription_list)
print(Review1_list)
print(Review2_list)          
```
