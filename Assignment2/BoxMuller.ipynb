```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as rd
```

```python
def f(theta,E):
  return np.cos(theta) * np.sqrt(2*E)
```

```python
def BoxMuller(N,mu=0,sigma=1):
  out = np.zeros(N)
  for i in range(N):
    theta = rd.uniform(low=0,high=2*np.pi,size=1)
    E = rd.exponential(scale = 1/1,size=1)
    X = f(theta,E)
    out[i] = mu+sigma*X
  return out
```

```python
def phi(X,mu,sigma):
  return 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(X-mu)**2/(2*sigma**2))
```

```python
%matplotlib inline
```

```python
X=BoxMuller(N=10**4,mu=5,sigma=2)
plt.hist(X,bins=50,density=True)
xx = np.linspace(-3,15,100)
plt.plot(xx,phi(xx,5,2))
```

```python
 # for mean =0,sigma=1
 X=BoxMuller(N=10**4,mu=0,sigma=1)
plt.hist(X,bins=50,density=True)
xx = np.linspace(-3,5,100)
plt.plot(xx,phi(xx,0,1))
```
