```python
import pandas as pd
import numpy as np
```

```python
#pd.set_option('display.max_rows',740)
file = pd.read_csv('case_time_series.csv')
file.drop(index=0,inplace=True)
file
```

```python
file['Date_YMD'] = pd.to_datetime(file["Date_YMD"],format="%Y-%m-%d")
#file['Date_YMD']
```

```python
filt = (file['Date_YMD'] >=pd.to_datetime('2021-05-01',format='%Y-%m-%d')) & (file['Date_YMD'] <=pd.to_datetime('2021-05-22',format='%Y-%m-%d'))
dataset_train =file.loc[filt]
dataset_train
```

```python
filt = (file['Date_YMD'] >=pd.to_datetime('23-05-2021',format='%d-%m-%Y')) & (file['Date_YMD'] <=pd.to_datetime('31-05-2021',format='%d-%m-%Y'))
dataset_test =file.loc[filt]
dataset_test
```

```python
m=22
# train death and days
days=np.arange(1,m+1)
daily_confirmed = dataset_train["Daily Confirmed"].to_numpy(dtype='float32')
daily_recovered = dataset_train["Daily Recovered"].to_numpy(dtype='float32')
deaths = np.abs(daily_recovered-daily_confirmed)
deaths,days
```

```python
A=np.sum(days)
B=np.sum(deaths)
C=np.sum(days*days)
D=np.dot(deaths,np.transpose(days))
```

```python
A,B,C,D
```

```python
w0 = (B * C- A * D) / (m * C - A * A)
w0
```

```python
w1 = (A * B - m * D) / (A * A - m * C)
w1
```

```python
test_days=np.arange(m+1,32)
test_deaths=np.abs(dataset_test['Daily Confirmed'].to_numpy(dtype='float32')-dataset_test['Daily Recovered'].to_numpy(dtype='float32'))
test_days,test_deaths
```

```python
### error on test set
linear_error=0
for i in range(0,9):
    linear_error += (w0+w1*days[i]-test_deaths[i])**2
print("root-mean-square error",np.sqrt(linear_error/9))
```

```python
# for 20April2021
d=-10
deaths1=w0+w1*d
print('predicted death',deaths1)
print('Actual death',np.abs(np.float32(file.loc[447,'Daily Confirmed'])-np.float32(file.loc[447,'Daily Recovered'])))
```

```python
# for 10june2020
d=41
deaths2=w0+w1*d
print('predicted death',deaths2)
print('Actual death',np.abs(np.float32(file.loc[498,'Daily Confirmed'])-np.float32(file.loc[498,'Daily Recovered'])))
```

```python
########### Quadratic model ########
#mat=numpy.concat(np.ones(m+1))
mat=np.stack((np.ones(m),days,days*days),axis=0)
mat = np.transpose(mat)
temp=np.dot(np.transpose(mat),mat)
temp1=np.dot(np.linalg.inv(temp),np.transpose(mat))
coff = np.dot(temp1,deaths)
w0=coff[0]
w1=coff[1]
w2=coff[2]
coff
```

```python
quadratic_error=0
for i in range(0,9):
    quadratic_error += (w0+w1*days[i]+w2*days[i]*days[i]-test_deaths[i])**2
print("root-mean-square error",np.sqrt(quadratic_error/9))
```

```python
# for 20April2020
d=-10
deaths1=w0+w1*d+w2*d*d
print('predicted death',deaths1)
print('Actual death',np.abs(np.float32(file.loc[447,'Daily Confirmed'])-np.float32(file.loc[447,'Daily Recovered'])))
```
