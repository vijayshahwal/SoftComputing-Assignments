```python
class_ = dataset['v1'].to_numpy()
data = dataset['v2'].to_numpy()
dataset = np.array([class_,data])
dataset = dataset.T
```

```python
def scaled_data(dataset):
    np.random.shuffle(dataset)
    dataset=dataset.T
    n=len(dataset.T)
    pct_data = int(0.7*n)
    x_train=dataset[1][:pct_data]
    y_train=dataset[0][:pct_data]
    x_test=dataset[1][pct_data:]
    y_test=dataset[0][pct_data:]
    return x_train,y_train,x_test,y_test
```

```python
def createDic(x_train,y_train):
    word=[]
    spam_word=[]
    ham_word=[]
    n=len(x_train)
    for i in range(n):
        word+= x_train[i].split(" ")
        if(y_train[i]=='spam'):
            spam_word += x_train[i].split(" ")
        else:
            ham_word += x_train[i].split(" ")
    ham_set = set(ham_word)
    spam_set = set(spam_word)
    word_set = set(word)
    dic_all_word = {}
    for i in word_set:
        dic_all_word[i]=word.count(i)
    dic_spam={}
    for i in spam_set:
        dic_spam[i]=spam_word.count(i)
    dic_ham={}
    for i in ham_set:
        dic_ham[i]=ham_word.count(i)
    return dic_all_word,dic_spam,dic_ham,word,spam_word,ham_word
```

```python
def prob_given_spam(w,dic_spam,total_spam):
    return dic_spam[w]/total_spam
```

```python
def prob_given_ham(w,dic_ham,total_ham):
    return dic_ham[w]/total_ham
```

```python
def prob_word(w,dic_all_word,total_word):
    return dic_all_word[w]/total_word
```

```python
def prob_spam(message,dic_spam,spam_word,dic_all_word,word):
    num = den =1
    for i in message.split():
        if(i in spam_word):
            num *= prob_given_spam(i,dic_spam,len(spam_word))
            den *= prob_word(i,dic_all_word,len(word))
    if(den==0):
        num+=1
        den+=1
    num*=len(spam_word)/len(word)
    return num/den
```

```python
def prob_ham(message,dic_ham,ham_word,dic_all_word,word):
    num = den =1
    for i in message.split():
        if(i in ham_word):
            num *= prob_given_ham(i,dic_ham,len(ham_word))
            den *= prob_word(i,dic_all_word,len(word))
    if(den==0):
        num+=1
        den+=1
    num*=len(ham_word)/len(word)
    return num/den
```

```python
def predict(message,dic_ham,dic_spam,ham_word,spam_word,word,dic_all_word):
    prob_spam_mess = prob_spam(message,dic_spam,spam_word,dic_all_word,word)
    prob_ham_mess = prob_ham(message,dic_ham,ham_word,dic_all_word,word)
    if(prob_spam_mess>prob_ham_mess):
        return 'spam'
    return 'ham'
```

```python
def accuracy(x_test, y_test, dic_all_word, dic_spam, dic_ham, word, spam_word, ham_word):
    count=0
    for i in range(len(y_test)):
        pred = predict(x_test[i],dic_ham,dic_spam,ham_word,spam_word,word,dic_all_word)
        if(y_test[i]==pred):
            count+=1
    return (count/len(y_test))*100
    
```

```python
x_train,y_train,x_test,y_test = scaled_data(dataset)
```

```python
dic_all_word,dic_spam,dic_ham,word,spam_word,ham_word = createDic(x_train,y_train)
print(accuracy(x_test,y_test,dic_all_word,dic_spam,dic_ham,word,spam_word,ham_word))
```
